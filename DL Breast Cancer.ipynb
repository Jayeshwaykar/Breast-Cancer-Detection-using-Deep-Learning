Separating the features and target

```python
X = Data_frame.drop(['label'],axis=1)
Y = Data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler =StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

Building Neural Network

```python
#importing tensorflow and keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
from keras.api._v2.keras import activations
model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)),# input layers
                          keras.layers.Dense(20, activation = 'relu'), # hidden layers
                          keras.layers.Dense(30, activation = 'sigmoid'),#hidden layers
                          keras.layers.Dense(2, activation = 'sigmoid') # output layers
])
```

```python
# compiling the neural network
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# training the neural network model
history = model.fit(X_train,Y_train,validation_split=0.1,epochs =20)
```

```python

# Visualizing accuracy and loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data','validation data'], loc = 'lower right')
```

Accuracy of the model on the test data

```python
loss, accuracy = model.evaluate(X_test_std,Y_test)
print(accuracy)
```
